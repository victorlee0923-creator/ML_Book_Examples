```python
import pandas as pd

df = pd.read_csv('winequality-red.csv', sep=';')
df.head()
```

```python
from sklearn.model_selection import train_test_split

X, y = df.iloc[:, :-1], df['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=0)
X_train.shape
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

models = {'DT': DecisionTreeClassifier(random_state=0), 
          'bagging': BaggingClassifier(DecisionTreeClassifier(), n_estimators=10, 
                                       max_samples=1., max_features=1., n_jobs=-1, random_state=0)}
scores = {}
for name, clf in models.items():
    clf.fit(X_train, y_train)
    scores[(name, 'Train score')] = clf.score(X_train, y_train)
    scores[(name, 'Test score')] = clf.score(X_test, y_test)
pd.Series(scores).unstack()
```
